```python
import pyodbc

# Exemple de chaîne de connexion plus courte avec un chemin réduit
conn_str = (
    r'DRIVER={Microsoft Access Driver (*.mdb, *.accdb)};'
    r'DBQ=./Gestion-DépotsBd.accde;'
)

# Connexion à la base de données
conn = pyodbc.connect(conn_str)
cursor = conn.cursor()

```

```python
for table_info in cursor.tables():
    print(table_info.table_name)
```

```python
# Exporter les données des tables Clients, Dépots, Informations_Société dans un fichier CSV avec le nom des colonnes
cursor.execute("SELECT * FROM Clients")
with open('Clients.csv', 'w') as f:
    f.write(','.join([column[0] for column in cursor.description]) + '\n')
    for row in cursor.fetchall():
        f.write(','.join(map(str, row)) + '\n')

cursor.execute("SELECT * FROM Dépots")
with open('Articles.csv', 'w') as f:
    f.write(','.join([column[0] for column in cursor.description]) + '\n')
    for row in cursor.fetchall():
        f.write(','.join(map(str, row)) + '\n')

cursor.execute("SELECT * FROM Informations_Société")
with open('Informations_Société.csv', 'w') as f:
    f.write(','.join([column[0] for column in cursor.description]) + '\n')
    for row in cursor.fetchall():
        f.write(','.join(map(str, row)) + '\n')

# Creation d'un fichier Depots.csv avec la clonne No_Depot, No_Article, Date_Dépot, EchéanceDépot de la table Depots
cursor.execute("SELECT No_Dépot, Date_Dépot, EchéanceDépot, NoClient FROM Dépots")
with open('Depots.csv', 'w') as f:
    f.write(','.join([column[0] for column in cursor.description]) + '\n')
    for row in cursor.fetchall():
        f.write(','.join(map(str, row)) + '\n')
        
```

```python
# des entête du fichier Client.csv (NoClient,NomClient,Prénom,Adresse 1,Adresse 2,Ville,CodePostal,Pays,Date Entrée,No Tél,No Mobile,No Prof,Remarques,Email) pour correspondre (id, nom, prénom, rue, ville, npa, dateCreation, mobile, numProf, remarque, email, telephone) 
with open('Clients.csv', 'r') as f:
    lines = f.readlines()
    lines[0] = 'id, nom, prénom, rue,Adresse 2, ville, npa,pays, dateCreation, telephone, mobile, numProf, remarque, email\n'
    with open('Clients.csv', 'w') as f:
        f.writelines(lines)


# des entête du fichier Articles.csv (No_Dépot,No_Article,Date_Dépot,Nature_Dépot,Taille,PrixClient,PrixVente,PrixVenteMin,EchéanceDépot,Date_Retour,Date_Réservation,Echéance_Réservation,MontantNégocié,AcompteVersé,Date_Vente,MontantVente,NoClient,Status_Dépot,No_Fiche_Dépots,NoClientOld) 
# pour correspondre depot_id, id, created_at,description,taille,prixClient,prixVente,prixSolde,EchéanceDépot,Date_Retour,Date_Réservation,Echéance_Réservation,MontantNégocié,AcompteVersé,Date_Vente,MontantVente,NoClient,Status_Dépot,No_Fiche_Dépots,NoClientOld\n
with open('Articles.csv', 'r') as f:
    lines = f.readlines()
    lines[0] = 'depot_id, id, created_at,description,taille,prixClient,prixVente,prixSolde,EchéanceDépot,Date_Retour,Date_Réservation,Echéance_Réservation,MontantNégocié,AcompteVersé,Date_Vente,MontantVente,NoClient,Status_Dépot,No_Fiche_Dépots,NoClientOld\n'
    with open('Articles.csv', 'w') as f:
        f.writelines(lines)
```
